```python
from matplotlib import pyplot as plt
```

```python
import os
import cv2

folder_path = "D:\\python\\conda\\null\\emotion\\Screenshots"
image_list = []

for filename in os.listdir(folder_path):
    if filename.endswith(".jpg") or filename.endswith(".png"):
        image_path = os.path.join(folder_path, filename)
        image = cv2.imread(image_path)
        image_list.append(image)

# Now you can use the image_list for further processing
```

plotting

```python
# Now you can use the image_list for further processing
# For example, to show the images
top_5_images = sorted(image_list, key=lambda x: x.size, reverse=True)[:5]

for image in top_5_images:
    plt.imshow(image)
    plt.show()
```

```python
num_images = int(input("Enter the number of images to print: "))
```

```python
printed_image_locations = []
print("Top", num_images, "image names:")
for image in top_5_images[:num_images]:
    image_path = os.path.join(folder_path, os.path.basename(image_path))
    printed_image_locations.append(image_path)
    print(image_path)
```

```python
image_variables = {}

for i, image_path in enumerate(printed_image_locations[:num_images]):
    variable_name = f"img{i+1}"
    image_variables[variable_name] = image_path

# Now you can access the image locations using the variable names
print(image_variables["img1"])  # prints the location of the first image
print(image_variables["img2"])  # prints the location of the second image
print(image_variables["img3"])  # prints the location of the third image
```

# Blur the image

```python
# Blur the image
fig_size = 7
blurred_img = cv2.blur(img, (fig_size, fig_size))

# Display the original and blurred images
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
ax[0].set_title('Original Image')
ax[0].axis('off')
ax[1].imshow(cv2.cvtColor(blurred_img, cv2.COLOR_HSV2RGB))
ax[1].set_title('Blurred Image')
ax[1].axis('off')
plt.show()
```

```python
# Blur the image
fig_size = 7
blurred_img = cv2.blur(img, (fig_size, fig_size))
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_HSV2RGB))
plt.title('Original Image')

plt.axis('off')

plt.subplot(122)
plt.imshow(cv2.cvtColor(blurred_img, cv2.COLOR_HSV2RGB))
plt.title('Blurred Image')
plt.xticks([]), plt.yticks([])

plt.show()

```

ZOOM

```python
# Display the original image
plt.imshow(img)
plt.title('Original Image')

# Set the zoom level
zoom_factor = 2  # Increase this value to zoom in more

# Calculate the new size of the image
new_width = image.shape[1] / zoom_factor
new_height = image.shape[0] / zoom_factor

# Set the zoomed-in region
plt.xlim(0, new_width)
plt.ylim(0, new_height)

# Show the zoomed-in image
plt.show()
```

ROTATE


```python
m=cv2.getRotationMatrix2D((img.shape[1]/2,img.shape[0]/2),40,1)
rotated_img=cv2.warpAffine(img,m,(img.shape[1],img.shape[0]))
plt.imshow(rotated_img)
```

```python
import numpy as np
```

TRANSLATE



```python
m = np.float32([[1, 0, 100], [0, 1, 50]])
translated_img = cv2.warpAffine(img, m, (img.shape[1], img.shape[0]))
plt.imshow(translated_img)
```

```python
original_img = cv2.cvtColor(img, cv2.COLOR_HSV2RGB)
translated = cv2.cvtColor(translated_img, cv2.COLOR_HSV2RGB)


# Create a figure with two subplots
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Plot the original image
axs[0].imshow(original_img)
axs[0].set_title("Original Image")

# Plot the translated image
axs[1].imshow(translated)
axs[1].set_title("Translated Image")

# Remove the axis labels
for ax in axs:
    ax.axis("off")

# Display the figure
plt.show()
```
